```python
from __future__ import print_function, division
from builtins import range,input

from keras.layers import Dense, Flatten, Lambda, Input
from keras.models import Model
from keras.applications.vgg16 import VGG16
from keras.applications.vgg16 import preprocess_input
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator

from sklearn.metrics import confusion_matrix
import numpy as np
import matplotlib.pyplot as plt

from glob import glob
```

```python
IMAGE_SIZE = [100,100]
EPOCHS = 5
BATCH_SIZE = 32
```

```python
image_files = glob('Training/*/*.jp*g')
validation_images = glob('Test/*/*.jp*g')
folders = glob('Training/*')
```

```python
plt.imshow(image.load_img(np.random.choice(image_files)))
plt.show()
```

# Start the model

```python
#include_top is for not taking the last dense layer to adapt the weights
vgg = VGG16(input_shape=IMAGE_SIZE + [3], weights='imagenet', include_top=False)
```

```python
for layer in vgg.layers:
    layer.trainable = False
```

```python
x = Flatten()(vgg.output)
prediction = Dense(len(folders), activation='softmax')(x)
```

```python
model = Model(inputs=vgg.input, outputs=prediction)
```

```python
model.summary()
```

```python
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

# Creating a Image data generator

```python
gen = ImageDataGenerator(
        rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    preprocessing_function=preprocess_input
)
```

```python
test_gen = gen.flow_from_directory('Test/', target_size=IMAGE_SIZE)
print(test_gen.class_indices)

labels = [None] * len(test_gen.class_indices)
for k, v in test_gen.class_indices.items():
    labels[v] = k 
```

```python
#vgg work with the color channel od BGR
for x, y in test_gen:
    print('min: ', x[0].min(), 'max: ', x[0].max())
    plt.title(labels[np.argmax(y[0])])
    plt.imshow(x[0])
    plt.show()
    break 
```

# Create generators

```python
train_generator = gen.flow_from_directory('Training/', target_size=IMAGE_SIZE, shuffle=True, batch_size=BATCH_SIZE)
valid_generator = gen.flow_from_directory('Test/', target_size=IMAGE_SIZE, shuffle=True, batch_size=BATCH_SIZE)
```

# Fit the model

```python
r = model.fit_generator(generator=train_generator, 
                        validation_data=valid_generator, 
                        epochs=EPOCHS,
                       steps_per_epoch=len(image_files) // BATCH_SIZE,
                       validation_steps=len(validation_images) // BATCH_SIZE)
```
